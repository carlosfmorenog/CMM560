## Loading data into a Jupyter Notebook 

Just as we loaded the wind farm image as data before, you must know that we can load data from different extensions (.csv, .tsv, .txt, .xlsx) to our notebooks!

The easiest way to do so is with a Python module called `Pandas`

```python
import pandas as pd
```

```python
# This cell will work only if the "data.csv" is in the same directory as this notebook
data = pd.read_csv('data.csv')
data
```

Commands such as `read_csv` can also take a web address as an input, and load images from online repos!

For instance, if `data.csv` is in my [Dropbox](https://www.dropbox.com/), I can copy the dropbox link, change `dl=0` for `raw=1` and load the file

```python
# This command will run anywhere (as long as you have internet connection)
data_dropbox = pd.read_csv('https://www.dropbox.com/s/my97tgyus8s05e1/data.csv?raw=1')
data_dropbox
```
